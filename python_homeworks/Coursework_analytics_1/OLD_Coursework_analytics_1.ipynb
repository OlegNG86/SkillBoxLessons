```python
#импортирую библиотеки
import pandas as pd
import numpy as np
import datetime as dt
import matplotlib.pyplot as plt



#открываю csv файлы
courses = pd.read_csv('./Data/courses.csv')
students = pd.read_csv('./Data/students.csv')
course_contents = pd.read_csv('./Data/course_contents.csv')
progresses = pd.read_csv('./Data/progresses.csv')
progress_phases = pd.read_csv('./Data/progress_phases.csv')
```

```python
# Объединяю все csv (кроме student.csv)

# Удалил ненужную колонку
del courses['Unnamed: 0']

# Переименовал названия колонок
courses.columns = ['id', 'course_title', 'course_field']

# Объединил 2 колонки
main = pd.merge(courses, course_contents, left_on='id', right_on='course_id')

# Удалил ненужную колонку
del main['id']

# Объединил 2 колонки
main_prog = pd.merge(progresses, progress_phases, left_on='id', right_on='progress_id')

# Удалил ненужную колонку
del main_prog['id']

# Объединил 2 колонки.
main = pd.merge(main, main_prog)
```

```python
# Выбрал колонки из примера, который был при запуске курсовой №1
main = main[['progress_id', 'module_number', 'lesson_number', 'status', 'start_date', 'finish_date', 'student_id', 'course_id', 'course_title', 'course_field', 'module_title', 'lesson_title', 'lesson_token', 'is_video', 'is_homework']]

# Отсортировал значения по колонке начала ДЗ, с применением к исходному датасету.
main.sort_values(by='start_date', inplace=True)

# Взял строчку из примера при запуске Курсовой №1
main[main.student_id == '768c2987a744c51ce64a5993a2a94eaf'] # проверка основного датасета
```

```python
#общее количество курсов в датасете
# Выбрал уникальные названия курсов и посчитал их количество
course_count = len(pd.unique(main['course_id']))
print(f'общее количество курсов в датасете {course_count}')
print()

#количество модулей на каждом курсе
# Сгруппировал датасет по названиям курсов и выбрал уникальные имена модулей.
module_count = main.groupby('course_title')['module_title'].nunique()
print(f'количество модулей на каждом курсе {module_count}')
print()

# количество уроков в каждом модуле на каждом курсе
# Сгруппировал датасет по названиям модулей и выбрал уникальные имена уроков.
lesson_count = main.groupby(['course_title', 'module_title'])['lesson_title'].nunique()
print(f'количество уроков в каждом модуле на каждом курсе {lesson_count}')
print()

# количество учеников на каждом курсе
# Сгруппировал датасет по названиям модулей и выбрал уникальные id студентов.
students_count_from_every_courses = main.groupby('course_title')['student_id'].nunique()
print(f'количество студентов на каждом курсе {students_count_from_every_courses}')
print()

# создаю датасет student_without_id_
# Выбрал из изначального датасета нужные мне колонки
student_without_id_ = students[['id', 'city', 'birthday']]

# колонку birthday перевожу в тип datetime
student_without_id_.loc[:, 'birthday'] = pd.to_datetime(student_without_id_.loc[:, 'birthday'], errors = 'coerce')

# создаю переменную now с текущей датой
now = pd.Timestamp('now')

# добавляю в датасет student_without_id_ колонку age с возрастом студентов
student_without_id_['age'] = (now - student_without_id_['birthday']).astype('<m8[Y]')

# Задал константы на минимальный и максимальный возраст
MIN_AGE = 5
MAX_AGE = 80

# Отфильтровал датасет по возрасту
student_without_id_ = student_without_id_[(student_without_id_['age'] > MIN_AGE) & (student_without_id_['age'] < MAX_AGE)]

# минимальный, максимальный, средний, медианный возраст студентов
min_age = student_without_id_['age'].describe()['min']
max_age = student_without_id_['age'].describe()['max']
mean_age = student_without_id_['age'].describe()['mean']
median_age = student_without_id_['age'].describe()['50%']
print(f'минимальный возраст студентов {min_age}')
print(f'максимальный возраст студентов {max_age}')
print(f'средний возраст студентов {mean_age}')
print(f'медианный возраст студентов {median_age}')

#переименовываю колоны student_without_id_
student_without_id_.columns = ['student_id', 'student_city', 'student_birthday', 'student_age']

#объединяю датасеты student_without_id_ и progresses
student_merge = pd.merge(student_without_id_, progresses)

# переименовываю колоны courses
courses.columns = ['course_id', 'course_title', 'course_field']

#объединяю датасеты student_merge и courses
student_merge = pd.merge(student_merge, courses)

# минимальный, максимальный, средний, медианный возраст студентов на каждом курсе
min_age_in_every_courses = student_merge.groupby('course_title')['student_age'].min()
max_age_in_every_courses = student_merge.groupby('course_title')['student_age'].max()
mean_age_in_every_courses = student_merge.groupby('course_title')['student_age'].mean()
median_age_in_every_courses = student_merge.groupby('course_title')['student_age'].median()

# Распечатываю аггрегированную таблицу где видно минимальный, максимальный, средний и медианные возрасты по каждому курсу.
student_merge.groupby('course_title')['student_age'].agg({'student_age' : ['min', 'max', 'mean', 'median']})
```

```python
# Медианное значение уроков в каждом модуле на каждом курсе.

# Сделал массив из курсов
course_array = main['course_title'].unique()

# Создал пустой словарь чтобы поместить в него на место ключей названия словарей, а значения ключей медианной количество.
median_lesson_count_in_every_courses = {}

# Перебираю в цикле названия курсов
for course in course_array:
    
    # Создаю массив модулей
    module_array = main[main['course_title'] == course]['module_title'].unique()
    
    # Создаю пустой список в который буду записывать медианные значения уроков в модулях на курсе
    lesson_count_in_module = []
    
    # Перебираю в цикле названия модулей
    for module in module_array:
        
        # Нахожу количество уроков в каждом модуле
        count = main[(main['course_title'] == course) & (main['module_title'] == module)]['lesson_title'].unique().size
        
        # Записываю значение полученное выше в список
        lesson_count_in_module.append(count)
        
    # В изначальный словарь записываю курс и медианное количество уроков в каждом модуле этого курса
    median_lesson_count_in_every_courses[course] = np.median(lesson_count_in_module)

# Смотрю получившийся словарь
median_lesson_count_in_every_courses

# P.S. Процесс долгий, но пришёл к нему т.к. не сообразил как сделать 
# это методами пандас, очень надеюсь что есть более быстрые пути.
```

```python
#Построил bar-chart, отражающий количество студентов на каждом курсе.
%matplotlib inline

# получил доступ к холсту
fig, subplot = plt.subplots()

# построил bar-chart
students_count_from_every_courses.plot.bar(rot=90)

# написал название для bar-chart
plt.title('Number of students in courses')

# отобразил график
plt.show()
```

```python
# (m) Постройте горизонтальный (столбцы должны располагаться горизонтально) bar-chart, 
# отражающий количество студентов на каждом курсе. График должен иметь заголовок. 
# Значения должны быть отсортированы. Цвет столбцов должен содержать информацию о сфере, 
# к которой относится курс (то есть нужна легенда). Прозрачность должна стоять на отметке 0.1. 
# На график должна быть нанесена линия медианы. У медианы должен быть свой цвет. Рамки у графика быть не должно ⭐

# Выбрал уникальные значения курсов из датасета main
courses_list = [i for i in main.groupby('course_title')['course_field'].unique().index]

# Сделал список из students_count_from_every_courses
count_student_list = list(students_count_from_every_courses)

# Сделал список категорий курсов
course_field_list = [x[0] for x in main.groupby('course_title')['course_field'].unique()]

# Сделал словарь состоящий из списков предыдущих пунктов
data = {'course':courses_list, 'field':course_field_list, 'student_count':count_student_list}

# Создал датасет из словаря предыдущего пункта
df = pd.DataFrame(data)

# Получил доступ к холсту
fig, ax = plt.subplots()

# Написал название графика
plt.title('Number of students in courses')

# Убрал рамки у графика
plt.box(on=False)

# Сгруппировал датафрейм по курсам и категориям, взял их сумму, выбрал колонку количество студентов, применил unstack()
df0 = df.groupby(['course','field']).sum()['student_count'].unstack()

# Добавил новую колонку в которую проссумировал все строки предыдущего датафрейма
df0['sorting'] = df0.sum(axis=1)

# Отсортировал по этой колонке с применением сортировки к исходному датафрейму.
df0.sort_values('sorting', inplace=True)

# Удалил более ненужную колонку сортировки
del df0['sorting']

# Построил график согласно условиям подпункта 4 из 1го задания.
ax = df0.plot(kind='barh', ax=ax, stacked=True, alpha=0.1)

# Нашёл медиану от количество учащихся студентов на разных курсах
median = df['student_count'].median()

# Нарисовал на графике медиану найденную в предыдущем пункте, у медианы свой цвет.
plt.axvline(median, color='k', linestyle='--', label='median', alpha=0.5)

# Подвинул легенду немного правее чтобы она не перекрывала график
ax.legend(bbox_to_anchor=(1.3, 0.75))

# Отобразил график
plt.show()
```

```python
# (p) Посчитать конверсию перехода студентов из одного модуля в другой на каждом курсе. 
# Формула: отношение количества студентов, приступивших к выполнению домашнего задания 
# в этом модуле (если дз в модуле несколько, то считать по первому дз в модуле), 
# к количеству студентов, сдавших задание в предыдущем модуле (если дз в модуле несколько, 
# то считать по последнему дз в модуле).

# Создаю сгруппированный датасет по названию курса, номеру и названию модуля, выбираю колонки с id студента, началом вып-я ДЗ и оконч. вып. ДЗ и считаю
grouped_count_students = filtered_main.groupby(['course_title', 'module_number', 'module_title'])[['student_id', 'start_date', 'finish_date']].count()

# Создаю пустой словарь
graph_dict = {}

# В цикле перебираю названия курсов.
for course in courses_list:
    
    # Создаю переменную в которую записываю датасет по каждому курсу с количеством студентов начавших и окончивших модуль.
    count_students = grouped_count_students.loc[course].reset_index()
    
    # Создаю переменную в которой считаю количество модулей в курсе
    course_length = len(count_students)
    
    # Создаю ключ с названием курса и значением пустой список
    graph_dict[course] = []
    
    # Произвожу итерацию по количеству модулей в курсе
    for i in range(course_length):
        
        # Устанавливаю условие что при первом элементе в итерации добавляется значение 100 в первый элемент пустого списка выше
        if i == 0:
            graph_dict[course].append(100)
            
        # Условие всех последующих итераций выше 0
        if i > 0:
            
            # Переменная количества студентов на начало прохождения первого модуля
            end_previous = count_students.loc[0]['start_date']
            
            # Количество студентов на начало каждого последующего модуля
            start_next = count_students.loc[i]['start_date']
            
            # Конверсия студентов на начало текущего модуля к изначальному количеству
            conversion = round(start_next / end_previous * 100, 2)
            
            # Наполнение списка значениями конверсий полученных выше.
            graph_dict[course].append(conversion)

# Создаётся датасет из словаря полученного выше в цикле.
df_graph = pd.DataFrame.from_dict(graph_dict, orient='index')

# Смотрю получившийся датасет
df_graph
```
