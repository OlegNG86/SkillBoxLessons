```python
import pandas as pd

df_courses = pd.read_csv('data/courses.csv')
df_course_contents = pd.read_csv('data/course_contents.csv')
df_progresses = pd.read_csv('data/progresses.csv')
df_progress_phases = pd.read_csv('data/progress_phases.csv')
df_students = pd.read_csv('data/students.csv')
```

```python
df_progresses
```

```python
# джойним датасеты так, чтобы было название курса, а не только айди
course_contents = df_course_contents.join(df_courses.set_index('id'), on='course_id')

# оставим нужные нам колонки
course_contents = course_contents[['title','module_number','lesson_number','module_title','lesson_title','is_homework']]

# выбираем модули, в которых есть домашняя работа
course_contents_hw = course_contents[course_contents['is_homework']==True]

# смотрим, где у нас ‘count’ в группировке будет больше единицы
course_contents_hw[course_contents_hw.groupby(['title','module_number'])['lesson_number'].transform('count') > 1]
```
